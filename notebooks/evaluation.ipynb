```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load evaluation metrics
with open('../datasets/cache/evaluation_metrics.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Extract metadata
metadata = data.pop('metadata', {})
print(f"Total qrels loaded: {metadata.get('total_qrels_loaded', 0)}")
print(f"Total queries used: {metadata.get('total_queries_used', 0)}")

# Display metrics dataframe
df = pd.DataFrame(data)
print("\nEvaluation Metrics:")
print(df.to_string(index=False))

# Plot charts
df.set_index('Model').plot(kind='bar', figsize=(10, 6))
plt.title('Evaluation Metrics by Model')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
